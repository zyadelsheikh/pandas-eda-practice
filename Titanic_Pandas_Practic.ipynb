## 5. Feature Engineering

```python
# Create new features using custom functions
def age_to_cat(age):
 if age < 4:
     return 0
 elif age < 21:
     return 1
 elif age < 33:
     return 3
 elif age < 50:
     return 4
 return 5 

age_cats={
    0:"baby",
    1:"child",
    2:"teen",
    3:"youngadult",
    4:"adult",
    5:"elder"

}
 
```

```python
df['age_phase']=df['age']
   
```

```python
df.head()
```

## 5. Feature Engineering

```python
# Create new features using custom functions
df['age_phase'] = df['age'].apply( age_to_cat )
```

```python
 for phase in df['age_phase'].unique():
     print(age_cats[phase])

     phase_df=df[df['age_phase']==phase]

     survived=phase_df[phase_df['survived']==1]

     perc=(survived.shape[0]/phase_df.shape[0])*100
     print("count: ",phase_df.shape[0])
     print("percentage :" ,perc)

     print("\n====\n")
```
